> Function to clean text data through pipeline

```python
import re
def remove_urls(text):
  """
  This function will try to remove URL present in out dataset and replace it with space using regex library.
  Input Args:
  text: strings of text that may contain URLs.
  Output Args:
  text: URLs replaces with text
  """
  url_pattern = re.compile(r'https?://\S+|www\.\S+')
  return url_pattern.sub(r'', text)
```

```python
def remove_emoji(string):
  """
  This function will replace the emoji in string with whitespace
  """
  emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
  return emoji_pattern.sub(r' ', string)
```

```python
def removeunwanted_characters(document):
  """
  This function will remove all the unwanted characters from the input dataset.
  Input Args:
  documet: A text data to be cleaned.
  Return:
  A cleaned document.
  """
  # remove user mentions
  document = re.sub("@[A-Za-z0-9_]+"," ", document)
  # remove hashtags
  document = re.sub("#[A-Za-z0-9_]+","", document)
  # remove punctuation
  document = re.sub("[^0-9A-Za-z ]", "" , document)
  #remove emojis
  document = remove_emoji(document)
  # remove double spaces
  document = document.replace('  ',"")
  return document.strip()
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.tokenize import RegexpTokenizer

def remove_punct(text):
  """
  This function removes the punctutations present in our text data.
  Input Args:
  text: text data.
  Returns:
  text: cleaned text.
  """
  tokenizer = RegexpTokenizer(r"\w+")
  lst=tokenizer.tokenize(' '.join(text))
  return lst
```

```python
def remove_stopwords(text_tokens):
  """
  This function removes all the stopwords present in out text tokens.
  Input Args:
  text_tokens: tokenize input of our datasets.
  Returns:
  result_tokens: list of token without stopword.
  """

  result_tokens = []
  for token in text_tokens:
    if token not in stop_words:
       result_tokens.append(token)
  return result_tokens
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize,pos_tag
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

def lemmatization(token_text):
  """
  This function performs the lemmatization operations as explained above.
  Input Args:
  token_text: list of tokens.
  Returns:
  lemmatized_tokens: list of lemmatized tokens.
  """
  lemma_tokens = []
  wordnet = WordNetLemmatizer()
  lemmatized_tokens = [wordnet.lemmatize(token, pos = 'v') for token in token_text]

  return lemmatized_tokens
```

```python
from nltk.stem import PorterStemmer

def stemming(text):
  """
  This function performs stemming operations.
  Input Args:
  token_text: list of tokenize text.
  Returns:
  stemm_tokes: list of stemmed tokens.
  """
  porter = PorterStemmer()
  stemm_tokens = []
  for word in text:
    stemm_tokens.append(porter.stem(word))
  return stemm_tokens
```

```python
def lower_order(text):
  """
  This function converts all the text in input text to lower order.
  Input Args:
  token_text : input text.
  Returns:
  small_order_text : text converted to small/lower order.
  """
  small_order_text = text.lower()
  return small_order_text
```

```python
# Function to remove numbers
def remove_numbers(text):
    return re.sub(r'\d+', '', text)
```

> **Building pipleline for text pre-processing**

```python
# Pipeline to clean the texts
def text_cleaning_pipeline(dataset, rule = "lemmatize"):

  # Lowercase all text to ensure uniformity.
  data = lower_order(dataset)

  # Remove URLs.
  data = remove_urls(data)

  # Remove mentions (@user), hashtags (#), emojis, and special characters.
  data = removeunwanted_characters(data)

  # Remove emojis.
  data = remove_emoji(data)

  # Remove numbers.
  data = remove_numbers(data)

  # Handle contractions (e.g., "don’t" → "do not").
  data = contractions.fix(data)

  # Create tokens.
  tokens = word_tokenize(data)

  # Remove stopwords.
  tokens = remove_stopwords(tokens)

  # Either lemmatize or stem words.
  if rule == "lemmatize":
    tokens = lemmatization(tokens)
  elif rule == "stem":
    tokens =stemming(tokens)
  else:
    print("Pick between lemmatize or stem")

  return " ".join(tokens)
```

> Use Keras Tokenizer to convert the text into sequences of integers.

```python
# Fit the tokenizer on the training data to learn the word indices
tokenizer.fit_on_texts(X_train)

# 3. Convert the texts to sequences of integers
X_train_sequences = tokenizer.texts_to_sequences(X_train)
X_val_sequences = tokenizer.texts_to_sequences(X_val)
# Confusion here read more properly
X_test_sequences = tokenizer.texts_to_sequences(X_test)
```

> Apply padding to ensure all sequences have the same length. Utilize percentile-based padding to
avoid excessively long sequences.

```python
# Calculate the length of each sequence
train_seq_lengths = [len(seq) for seq in X_train_sequences]

# Choose length where, say, 95% of your sequences fall under.
max_len = int(np.percentile(train_seq_lengths, 95))
print(f"Max Sequence Length (95th percentile): {max_len}")

# 5. Apply padding to ensure uniform sequence length
X_train_padded = pad_sequences(X_train_sequences, maxlen=max_len, padding='post')
X_val_padded = pad_sequences(X_val_sequences, maxlen=max_len, padding='post')
X_test_padded = pad_sequences(X_test_sequences, maxlen=max_len, padding='post')

# Now you can use `train_padded` and `test_padded` for further processing or model training
print("Training data shape:", X_train_padded.shape)
print("Testing data shape:", X_val_padded.shape)
print("Testing data shape:", X_test_padded.shape)
```

### **LSTM** with pretrained Word2Vec embeddings.

```python
embedding_dim = 50 # Match with downloaded GloVe model dimension
embedding_matrix = np.zeros((vocab_size, embedding_dim))
for word, i in tokenizer.word_index.items():
  if word in embedding_model:
    embedding_vector = embedding_model[word]
    embedding_matrix[i] = embedding_vector
  else:
    # Words not found in embedding index will be all-zeros
    pass

# Build the model
transfer_model = Sequential()
transfer_model.add(Embedding(
    input_dim=vocab_size,
    output_dim=embedding_dim,
    weights=[embedding_matrix],
    input_length=max_len,
    trainable=False,
    embeddings_regularizer=l2(0.001)
))
transfer_model.add(LSTM(64, return_sequences=False, kernel_regularizer=l2(0.001)))
transfer_model.add(Dropout(0.5))
transfer_model.add(BatchNormalization())
transfer_model.add(Dense(1, activation='sigmoid'))

transfer_model.compile(optimizer='adam',
                       loss='binary_crossentropy',
                       metrics=['accuracy'])

transfer_model.summary()
```

```python
checkpoint = ModelCheckpoint("transfer_learning.h5",
                             monitor="val_loss",
                             save_best_only=True,
                             mode="min",
                             verbose=1)

start_time = time.time()
transfer_model_history = transfer_model.fit(
    X_train_padded,
    y_train,
    epochs=30,
    batch_size=64,
    validation_data=(X_val_padded, y_val),
    callbacks=[early_stop,checkpoint],
    # class_weight=class_weights
)
transfer_model_training_time = time.time() - start_time
transfer_model_memory = get_memory_usage()
```
